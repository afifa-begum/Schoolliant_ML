```python
import numpy as np
```

```python
np.random.seed(42)
subjects = ["Math", "Physics", "Chemistry", "Biology", "English"]
marks = np.random.randint(0, 101, size=(100, 5))
```

```python
total_marks = np.sum(marks, axis=1)

average_marks = np.mean(marks, axis=1)

top_scorer = np.argmax(total_marks)

lowest_scorer = np.argmin(total_marks)

print("Top Scorer:")
print("Student", top_scorer+1)
print("Total =", total_marks[top_scorer])

print("\nLowest Scorer:")
print("Student", lowest_scorer+1)
print("Total =", total_marks[lowest_scorer])
```

```python
print("\nSubject Analytics")

for i in range(5):

    print(subjects[i])

    print("Average =", np.mean(marks[:,i]))

    print("Highest =", np.max(marks[:,i]))

    print("Lowest =", np.min(marks[:,i]))

    print()
```

```python
print("Students scoring above 80 average:")

print(np.where(average_marks > 80)[0] + 1)


print("\nStudents failing:")

fail_students = np.where(np.any(marks < 35, axis=1))[0] + 1

print(fail_students)


print("\nStudents scoring above 90 in any subject:")

above90 = np.where(np.any(marks > 90, axis=1))[0] + 1

print(above90)
```

```python
rankings = np.argsort(total_marks)[::-1]

print("\nTop 3 Students")

for i in range(3):

    print("Rank", i+1)

    print("Student", rankings[i]+1)

    print("Total", total_marks[rankings[i]])

    print()
```
